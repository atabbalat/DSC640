```python
# Imports
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
%matplotlib inline
```

```python
# Settings
pd.set_option("display.max_rows", 10, "display.max_columns", 10)
plt.style.use('seaborn')
sns.set_style('whitegrid')
```

#### Data Imports

```python
# Import hotdog dataset
df_hotdog = pd.read_excel('hotdog-contest-winners.xlsm')
df_hotdog.columns = ['Year', 'Winner', 'Dogs_eaten', 'Country', 'New_record']
df_hotdog
```

```python
df_hotdog.dtypes
```

```python
# Import Obama dataset
df_obama = pd.read_excel('obama-approval-ratings.xls')
df_obama
```

### Applying Charts

#### Bar Chart

```python
fig, ax = plt.subplots(figsize=(20, 10))

x = df_hotdog['Year']
y = df_hotdog['Dogs_eaten']
hue = df_hotdog['New_record']

# Plot
sns.barplot(x=x, y=y, hue=hue, palette=['navy', 'orange'], dodge=False)

# Format
plt.xlabel('Year', fontsize=15)
plt.xticks(fontsize=13)
plt.ylabel('Hotdogs Eaten', fontsize=15)
plt.yticks(fontsize=13)
plt.title('Python - Bar Chart\nHotdogs Eaten vs Year', fontsize=25)
plt.legend(title='New Record', fontsize=15)
```

#### Stacked Bar Chart

```python
fig, ax = plt.subplots(figsize=(20, 10))

x1 = df_obama['Issue']
y1 = df_obama['Approve']
y2 = df_obama['Disapprove']
y3 = df_obama['None']

# Plot
sns.barplot(x=y1, y=x1 , palette=['navy'], dodge=False, orient='h')
sns.barplot(x=y2, y=x1, left=y1, palette=['orange'], dodge=False, orient='h')
sns.barplot(x=y3, y=x1, left=y1+y2 ,palette=['grey'], dodge=False, orient='h')

# Format
plt.xlabel('Issue', fontsize=15)
plt.xticks(fontsize=13)
plt.ylabel('Rating', fontsize=15)
plt.yticks(fontsize=13)
plt.title('Python - Stacked Bar Chart\nObama Approval Ratings', fontsize=25)
```

#### Pie Chart

```python
fig, ax = plt.subplots(figsize=(20, 10))

x2 = df_hotdog['Country'].value_counts()
colors = sns.color_palette('pastel')

# Plot
plt.pie(x2.values, labels=x2.index, colors=colors, autopct='%.0f%%')

# Format
plt.title('Python - Pie Chart\nPercentage Wins Per Country', fontsize=25)
```

#### Donut Chart

```python
fig, ax = plt.subplots(figsize=(20, 10))

# Original Pie
x2 = df_hotdog['Country'].value_counts()
colors = sns.color_palette('pastel')

# Plot
plt.pie(x2.values, labels=x2.index, colors=colors, autopct='%.0f%%')

# Mid Circle
center = plt.Circle((0, 0), 0.7, fc='white')
fig = plt.gcf()

# Add Circle to pie
fig.gca().add_artist(center)

# Format
plt.title('Python - Donut Chart\nPercentage Wins Per Country', fontsize=25)
```
