```python
# Library Imports
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
%matplotlib inline
```

```python
# Settings
pd.set_option("display.max_rows", 10, "display.max_columns", 10)
plt.style.use('seaborn')
sns.set_style('whitegrid')
```

```python
# Data import
df = pd.read_excel('world-population.xlsm')
df
```

```python
df['Population1000'] = df['Population'] / 1000000

x = df['Year']
y = df['Population1000']
```

### Applying Charts

#### Line Chart

```python
fig, ax = plt.subplots(figsize=(20, 10))

# Plot
sns.lineplot(x=x, y=y, color='orange')

# Format
plt.xlabel('Year', fontsize=15)
plt.xticks(fontsize=13)
plt.ylabel('World Population in Millions', fontsize=15)
plt.yticks(fontsize=13)
plt.title('Python - Line Chart\nWorld Population in Millions', fontsize=25)
```

#### Step Chart

```python
fig, ax = plt.subplots(figsize=(20, 10))

# Plot
plt.step(x, y, color='orange')

# Format
plt.xlabel('Year', fontsize=15)
plt.xticks(fontsize=13)
plt.ylabel('World Population', fontsize=15)
plt.yticks(fontsize=13)
plt.title('Python - Step Chart\nWorld Population', fontsize=25)
```
